```python
def find_outliers(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   ot = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return ot
```

```python
ot = find_outliers(df[["pkts","dport"]])

ot


  
```
